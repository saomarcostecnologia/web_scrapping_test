```python
#Importar bibliotecas necessarias
from selenium import webdriver  
from selenium.webdriver.common.keys import Keys
from time import sleep 
import pyautogui as py
import pandas as pd

# 1 importar  web driver
#driver.maximize_window()
driver = webdriver.Chrome()

# 2 Através do site do banco central pegar cotações dia de Euro e Dolar(eua)
driver.get('https://www.google.com.br/')
sleep(3)
driver.find_element('xpath','//*[@id="APjFqb"]').send_keys("cotação dolar")
driver.find_element('xpath','//*[@id="APjFqb"]').send_keys(Keys.ENTER)
cota_dolar = driver.find_element('xpath',
                                 '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]'
                                 ).get_attribute('data-value')
driver.get("https://www.google.com/")
driver.find_element('xpath','//*[@id="APjFqb"]').send_keys("cotação Euro")
driver.find_element('xpath','//*[@id="APjFqb"]').send_keys(Keys.ENTER)
cota_euro = driver.find_element('xpath',
                                '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]'
                                ).get_attribute('data-value')
driver.get("https://www.melhorcambio.com/ouro-hoje")
cota_ouro = driver.find_element('xpath',
                                '//*[@id="comercial"]'
                                ).get_attribute('value')
cota_ouro = cota_ouro.replace(",",".")
driver.quit()

# 3 Importar base de dados Venda.xlsx
tabela = pd.read_excel(r'Produtos.xlsx')
display(tabela)
# 4 Fazer conversao de valores de produtos de real para dolar e euro
tabela.loc[tabela["Moeda"] == "Dólar","Cotação"] = float(cota_dolar)
tabela.loc[tabela["Moeda"] == "Euro","Cotação"] = float(cota_euro)
tabela.loc[tabela["Moeda"] == "Ouro","Cotação"] = float(cota_ouro)
display(tabela)

# 5 informar valor total gasto em euro e dolar

# 5.1 Novo preço compra depois novo preeço venda

tabela["Preço de Compra"] = tabela["Cotação"] * tabela["Preço Original"]
tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela["Margem"]
#Regra de formatação fazer após todos os calculos serem feitos
#tabela["Preço de Venda"] = tabela["Preço de Venda"].map("R${:.2f}".format)
#tabela["Preço de Compra"] = tabela["Preço de Compra"].map("R${:.2f}".format)
display(tabela)

# 5.2 Soma total em euro,dolar e ouro
# Calcular o total isolado de cada moeda
total_por_moeda = tabela.groupby("Moeda")["Preço de Compra"].sum()
media_por_moeda = tabela.groupby("Moeda")["Preço de Compra"].median()
#exemplos do que é possivel fazer com o pyhton semelhante so sql 

print(total_por_moeda)
print(media_por_moeda)


```
